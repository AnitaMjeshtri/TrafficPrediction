```python
import pandas as pd
import numpy as np
import seaborn as sns
import os

import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from sklearn.base import BaseEstimator, TransformerMixin, ClassifierMixin, clone
from sklearn.model_selection import KFold
from scipy import stats
from scipy.stats import norm, skew 
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.feature_selection import SelectFromModel
import warnings
warnings.filterwarnings('ignore')

def createHistogram(data, feature):
    sns.histplot(data=data,x=feature,hue='Traffic Situation',kde=True)
    plt.title('Histogram of ' + feature)
    plt.show()

def createSwarmPlot(data, feature):
    sns.swarmplot(x=feature,y='Traffic Situation',data=data)
    plt.title('Swarmplot of ' + feature)
    plt.show()

def createBoxPlot(data, feature):
    sns.boxplot(x=feature,y='Traffic Situation',data=data)
    plt.title('Boxplot of ' + feature)
    plt.show()

def createCatPlot(data, feature):
    sns.catplot(x=feature,y='Traffic Situation',data=data, kind='boxen')
    plt.title('Catplot of ' + feature)
    plt.show()


#* Read data from csv file
traficData = pd.read_csv('TrafficTwoMonth.csv')
print(traficData, "\n")

train, test = train_test_split(traficData, test_size=0.1, random_state=1992)
print("Shape of train: " , train.shape)
print("Shape of test: ", test.shape, "\n")

#* Check for missing values
print("Missing values in train data: ", train.isnull().sum().sum())
print("Missing values in test data: ", test.isnull().sum().sum(), "\n")
    
#? Time Visualization
createHistogram(train, 'Time')
createSwarmPlot(train, 'Time')
createBoxPlot(train, 'Time')
createCatPlot(train, 'Time')

#? Date Visualization
createHistogram(train, 'Date')
createSwarmPlot(train, 'Date')
createBoxPlot(train, 'Date')
createCatPlot(train, 'Date')

#? Day of the week Visualization
createHistogram(train, 'Day of the week')
createSwarmPlot(train, 'Day of the week')
createBoxPlot(train, 'Day of the week')
createCatPlot(train, 'Day of the week')

#? Count of Cars Visualization
createHistogram(train, 'CarCount')
createSwarmPlot(train, 'CarCount')
createBoxPlot(train, 'CarCount')
createCatPlot(train, 'CarCount')

#? Count of Bikes Visualization
createHistogram(train, 'BikeCount')
createSwarmPlot(train, 'BikeCount')
createBoxPlot(train, 'BikeCount')
createCatPlot(train, 'BikeCount')

#? Count of Buses Visualization
createHistogram(train, 'BusCount')
createSwarmPlot(train, 'BusCount')
createBoxPlot(train, 'BusCount')
createCatPlot(train, 'BusCount')

#? Count of Trucks Visualization
createHistogram(train, 'TruckCount')
createSwarmPlot(train, 'TruckCount')
createBoxPlot(train, 'TruckCount')
createCatPlot(train, 'TruckCount')

#? TOTAL Visualization
createHistogram(train, 'Total')
createSwarmPlot(train, 'Total')
createBoxPlot(train, 'Total')
createCatPlot(train, 'Total')
```
